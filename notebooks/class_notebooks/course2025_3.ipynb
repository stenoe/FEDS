# Let's recreate the quadratic function using a dataframe

1. create x axis data via a loop
2. set up a dataframe and store the x axis into it
3. create a new column and calculate the square function

```python
# 1. Generate the x-axis data by a loop

xdata = [] # empty list

for i in range(-50, 50+1, 1): # have data running from -50 to 50
    xdata.append(i)

# print first three and last three entries
print(f"xdata: {xdata[:3]} ... {xdata[-3:]}")
```

```python
# set up datframe

df1 = pd.DataFrame({"X" : xdata})
#df1 = pd.DataFrame(xdata, columns=["X"])
df1
```

```python
# 3. Create a new column which shows the square of X (x2)

df1["x2"] = df1.X * df1.X
df1
```

```python
# show first rows of the data
df1.head(3)
```

## Plotting the dataframe

```python
# Plot only the square function
df1.x2.plot();
```

# Create scatter plots

```python
# create a scatter plot using X and Y data

df1.plot.scatter(x="X", y="x2");
```

## Add a qubic function $y = x^3$ to the dataframe

```python
df1["x3"] = df1.x2 * df1.X
df1
```

```python
df1.plot.scatter(x="X", y="x3");
```

# Plotting with matplotlib

```python
# import the matplotlib library

import matplotlib.pyplot as plt

# Set a figure size and store it in a ax object
fig, ax = plt.subplots(1, 1, figsize=(5,5))

# Plot the data
# ax.plot(df1.X, df1.x2)
ax.scatter(df1.X, df1.x3)

# increase font size for the plot
plt.rc("font", size=14)

# make axis decorations
ax.set_xlabel("x")
ax.set_ylabel("$y = x^3$")
ax.set_title("Qubic function")

# show the plot
plt.show()
```
