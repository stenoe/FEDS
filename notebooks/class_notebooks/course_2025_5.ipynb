```python
# read in the data
fname = "land_global_temperature_anomaly_1850-2023.csv"
year, temp_anomaly = np.loadtxt(fname, delimiter=",", skiprows=5, unpack=True)
```

## make a linear regression on the temperature anomaly data

x = year

y = temp. anomaly data

Regression linear function $y = a_1 x + a_0$

coefficients are:

$$ a_1 = \frac{\sum_{i=0}^{n} y_i(x_i - \overline{x})}{\sum_{i=0}^{n} x_i(x_i - \overline{x})}, a_0 = \overline{y} - a_1 \overline{x} $$



```python
# calculate the mean values

year_mean = np.mean(year)
temp_anomaly_mean = np.mean(temp_anomaly)

# build coefficient a1
a1 = np.sum(temp_anomaly * (year - year_mean)) / np.sum(year*(year - year_mean) )

# build coefficient a0
a0 = temp_anomaly_mean - a1 * year_mean
```

```python
print(a1, a0)
```

```python
reg = a0 + a1 * year

# change figure size
plt.figure(figsize=(10, 5))

#plotting
plt.plot(year, temp_anomaly)
plt.plot(year, reg, "k--", linewidth=2, label="Regression line")
plt.legend()
plt.title("Global temperature anomaly")
plt.xlabel("Year")
plt.ylabel("Land temperature anomaly [C]")
plt.grid();
```

## Regression using the numpy polyfit function

```python
#calculate coefficients with polyfit

a1_new, a0_new = np.polyfit(year, temp_anomaly, deg=1)

linear = np.poly1d([a1_new, a0_new])

print(a1_new, a0_new)
print(a1, a0)
```

```python
# plotting new result

plt.figure(figsize=(10, 5))

#plotting
plt.plot(year, temp_anomaly)
plt.plot(year, linear(year), "k--", linewidth=2, label="Regression line")
plt.legend()
plt.grid();
```

```python
# create two datasets year1 and year2 and same for temp. anomaly
year1, temp_anomaly1 = year[0:120], temp_anomaly[0:120]
year2, temp_anomaly2 = year[120:], temp_anomaly[120:]

m1, b1 = np.polyfit(year1, temp_anomaly1, deg=1)
m2, b2 = np.polyfit(year2, temp_anomaly2, deg=1)

f_linear1 = np.poly1d([m1, b1])
f_linear2 = np.poly1d([m2, b2])

# plot the result
plt.figure(figsize=(10, 5))
plt.plot(year, temp_anomaly)
plt.plot(year1, f_linear1(year1), "g--", label="1850-1970")
plt.plot(year2, f_linear2(year2), "r--", label="1970-2023")
plt.ylabel("Global temperature anomaly [C]")
plt.xlabel("Year")
plt.grid()
plt.legend();
```

```python
print(f"1850 - 1970 slope:{m1}, intercept:{b1}")
print(f"1970 - 2023 slope:{m2}, intercept:{b2}")
```
