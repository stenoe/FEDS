## Accessing list elements

The AI generated a very compact code, it includes an iterator using a `for` loop. Then multiplies the element with itself and retuns a "new" list.

```python
[x * x for x in list1]
```

A more "traditional" approach is to create first an emtpy list, then access list elements usinf again a `for` loop wit a variable `x` that holds the current list element, append the square of the element to the new list.

```python
list2 = []
for x in list1:
  list2.append(x * x)

list2
```

A third way is to use an index `i` that iterates over the length of the list and "add" the squared elements that we "address" using the index in `[i]` brackets.

```python
list3 = []
for i in range(len(list1)):
    list3 = list3 + [list1[i] * list1[i]]

list3
```

## Let's plot the resulting list

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(list1, list2, marker='o', linestyle='-')
plt.xlabel('list1')
plt.ylabel('list2')
plt.title('Plot of list2 against list1')
plt.grid(True)
plt.show()
```
