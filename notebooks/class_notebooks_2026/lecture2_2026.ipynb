## Filling a list programatically

We use a `for` loop to access all possible elements in a list or we can append new elements with a `range` of numbers.


### Excursion: how to use the range function

Here we use the `range` function that allows us to either the number of wanted entries which then start automatically with 0.

````python
for x in range(5):
  print(x)
````

This may not what we wanted, we got five elements but if we wanted to start from let's say 1 and end at 5 we need to tell the function this by putting the appropriate parameters

````python
for x in range(1, 6):
  print(x)
````

Bear in mind, that the `range` function has a signature of

```python
range(start, stop, step)
```

where stop is the number of elements you you want + 1!

A way to make it clear is to use a variable that keeps the number of elements wanted and add the 1.

````python
n = 5 # number of elements in the range
for x in range(1, n+1):
  print(x)
````

## Now fill the list

We want it to hold the numbers from -50 to 50 in steps of 2.

````python
data1 = []  # empty list

# start = -50
# stop = 50 + 1
# step = 2
for i in range(-50, 51, 2):
  data1.append(i) # append to the list

print(data1) # print the list
````

Lets calcualate the square $y = x * x = x^2$

````python
data1 * data1
````

Oops, that produces an error! We can not simply multiply the lists because it is not clearly defined how to do that. Lists can hold any other datatype and we can not assume that there are only numerical values inside.

What we actually want is that we multiply every element in the list by itself!

Solution 1: use another loop for the calculation


````python
result1 = [] # new empty list

for i in range(len(data1)):
  result1.append(data1[i] * data1[i])

print(result1)
````

## How can we make the calculation more as we know from math lessons

We can use the numpy library and transfer the list inro an array (or better vector or matrix) that is limited to numerical values and has the proper operations defined.

We use first a conveniece function `numpy.square` that will calculate the square of the argument which can be a list or array.

````python
import numpy as np

result = np.square(data1)

result
````

````python
type(result)
````

````python
result.shape
````

Here, we use the `numpy.array` function to create a vector from the list.

````python
data2 = np.array(data1)
data2
````

````python
# now we can use the "normal" multiplication symbol

data2 * data2
````
