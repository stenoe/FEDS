# Dealing with tabular data

## Dataframes

see info on dataframes provided by the Pandas package here: https://pandas.pydata.org/

```python
# import the pandas dataframe lib
import pandas as pd

# create a dictionary with some data
data = {
    "calories" : [420, 380, 390],
    "duration" : [50, 40, 45]
}

# load into the a dataframe
df = pd.DataFrame(data)

# print the dataframe
print(df)
```

Access the elements

1. Access rows
2. Access columns
3. Access elements

```python
# get the first row

df.iloc[0]
```

```python
# get a column

df["calories"]
```

```python
# shorter way to access the column

df.calories
```

```python
# access using the `loc` function

df.loc[:, "duration"]
```

```python
# acces column by index

df.iloc[:, 1]
```

```python
# accessing elements by index

df.iloc[0, 1]
```

```python
# get some info of the dataframe
print(f"My dataset has {df.size} elements and its shape is {df.shape} and it has {df.ndim} columns")
```

## Make the example with the square from last lecture with a dataframe

1. create a x-axis data via a loop or range
2. set up the dataframe an store the x-axis data into it
3. create a new column with the square of the x-axis data

Here, the Colab built-in AI has been used!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a data array ranging from -50 to 50 in steps of 2
x_data = np.arange(-50, 51, 2)

# Add this to a new dataframe as column "x"
df_new = pd.DataFrame({"x": x_data})

# Create a new column with the squared values of the x column
df_new["x_squared"] = df_new["x"]**2

# Print the new dataframe (optional, for verification)
print(df_new)

# Plot the dataframe
df_new.plot(x="x", y="x_squared", title="x vs x_squared");
plt.xlabel("x")
plt.ylabel("x_squared")
plt.grid(True)
plt.show()
```
